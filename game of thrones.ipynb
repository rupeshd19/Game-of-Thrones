```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
battle=pd.read_csv('Downloads/battles.csv')
```

```python
battle.head()
```

```python
battle.shape
```

# rename column  attacker_1 to  primary_attacker 

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
```

```python
battle.head()
```

# rename defender_1 column to primary_defender

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
```

```python
battle.head()
```

# how many times a king has attacked

```python
battle['attacker_king'].value_counts()
```

```python
battle['defender_king'].value_counts()
```

# the locations where most number of times batlle fought

```python
battle['location'].value_counts()
```

# plot showing the size of army of king during he attacked

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

# plot showing the size of army of king when he got attcked by someone else 

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

#  plot showing king's battle type

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
